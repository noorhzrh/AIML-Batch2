    <h3> Why Do We Need Calculus in Machine Learning?


In Machine Learning, we often:

- Have a model with **parameters** (weights)
- Measure how good/bad the model is using a **loss/cost function**
- Want to **adjust** the parameters so the loss becomes **smaller**

Example:

- When we predict house price from size:
  
  y' = w . x + b
  
- We define a **loss** = “how far are our predictions from true values?”
- We want to **find values of w and b** that **minimize** this loss

To know **how to change w and b**, we need the idea of:

> “If I move w a little, does the loss go up or down?”  

This is exactly what **derivatives** and **gradients** tell us.

<div style="background-color: lightgreen; color: black; padding: 4px;">
    <h3> Derivative



A **derivative** measures the **rate of change** of a function.

Think of:

- Distance vs time → derivative is **speed**
- Height vs time → derivative could be **vertical speed**

For a function \( f(x) \):

- The derivative \( f'(x) \) tells us **how fast f(x) is changing** when x changes.
- More precisely: it is the **slope of the curve** at point x.

### Slope on a Line (Simple Case)

For a straight line:

\[
y = 2x + 3
\]

- Slope is constant = 2  
  This means: if x increases by 1, y increases by 2.

slope = x2​−x1​ / y2​−y1​
- How much does y change when x changes a little?

### Slope on a Curve (More Interesting)

For a curve like:

\[
f(x) = x^2
\]

- The slope is different at different points.
- Around x = 0, the curve is flatish.
- Around x = 2, the curve is steeper (going up faster).

The derivative tells us that slope:

- At x = 0, derivative of \( x^2 \) is 0
- At x = 2, derivative is 4
- At x = -2, derivative is -4 (going down there)


> The derivative at a point = slope of the curve at that point.

<div style="background-color: lightgreen; color: black; padding: 4px;">
    <h3> Gradient Descent

Imagine you are standing on a **hill** in the dark and want to reach the **lowest point** (valley).

You can’t see the whole hill, but you can feel which direction is **downwards**.

Gradient descent works like this:

1. Start at some **random point**
2. Look at the **slope** (derivative):
   - If slope is **positive** → function is going up as x increases → so move x **left** (decrease x)
   - If slope is **negative** → function is going down as x increases → so move x **right** (increase x)
3. Take a small step in the **downhill** direction
4. Repeat many times until you reach near the **minimum**

In ML:

- The “hill” is the **loss function**
- The “position” is the **parameter** (like w)
- Gradient descent adjusts w to **reduce loss**

    <h4> A Simple Example Function

Let’s use a simple function:

\[
J(w) = (w - 3)^2
\]

- This is a parabola (U-shaped curve)
- It reaches its **minimum** at \( w = 3 \)
- Our goal: **start from random w** and use **gradient descent** to move towards 3

We’ll:

1. Plot this function  
2. Approximate its derivative  
3. Apply gradient descent to see w move towards 3

```python
# 5. Import libraries

import numpy as np
import matplotlib.pyplot as plt

# Make plots look a bit nicer
plt.rcParams["figure.figsize"] = (7, 5)
```

```python
# 5.1 Define a simple function f(x) = x^2

def f(x):
    return x**2

# Create x values
x_vals = np.linspace(-5, 5, 200)  # 200 points from -5 to 5
y_vals = f(x_vals)

# Plot the function
plt.plot(x_vals, y_vals)

plt.axhline(0, color="red", linewidth=0.5)
plt.axvline(0, color="red", linewidth=0.5)

plt.title("f(x) = x^2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
```

    <h4> Approximating the Derivative Numerically

Instead of using formulas, we can **approximate** the derivative using a small step `h`:

\[
f'(x) / {f(x + h) - f(x)}{h}
\]

We’ll write a small function in Python to do this.

Small steps:
- 1e-5 = 0.00001
- 1e-1 = 0.1

```python
def numerical_derivative(func, x, h=1e-5):

    """
    Approximate the slope (derivative) of func at position x.

    We do this by:
    - Taking two very close x-values: x and x + h
    - Looking at how much func(x) changes between these two points
    - Dividing "change in y" by "change in x"
    """
    y1 = func(x)        # value at x
    y2 = func(x + h)    # value at x + small step
    return (y2 - y1) / h
```

```python
# Numerical derivative function

def numerical_derivative(func, x, h=1e-5):
    """
    Approximate derivative of func at point x using a small step h.
    """
    return (func(x + h) - func(x)) / h

# Test derivative of f(x) = x^2 at different points
for point in [-2, -1, 0, 1, 2]:
    slope = numerical_derivative(f, point)
    print(f"Approx derivative at x = {point}: {slope:.3f}")
```

#### Observation

For \( f(x) = x^2 \):

- At x = 0 → slope ≈ 0  
- At x = 1 → slope ≈ 2  
- At x = 2 → slope ≈ 4  
- At x = -2 → slope ≈ -4  

This matches the **exact derivative** \( f'(x) = 2x \).

We didn’t need the formula; we used a **numerical approximation**.

To estimate the slope (derivative) at a point,
we move a tiny step to the right, see how much the function’s value changes,
and divide that change by the size of our step.

<div style="background-color: lightgreen; color: black; padding: 4px;">
    <h3> Our Cost Function for Gradient Descent

In machine learning, we often have a **parameter** (or many parameters) that we want to learn.  
Here we call our parameter **`w`**.

A **cost function** (also called **loss function**) tells us:

> “How bad is this value of `w`?”

- If the cost is **small** → `w` is good.
- If the cost is **large** → `w` is bad (we need to change it).

Now let’s work with:

\[
J(w) = (w - 3)^2
\]

We know the minimum is at \( w = 3 \), but we’ll pretend we don’t know and use **gradient descent** to find it.



#### Derivative of J(w)

We won’t force heavy calculus, but for this function:

\[
J'(w) = 2(w - 3)
\]

We can:

- Either use this exact formula, or  
- Use our numerical derivative function from before

We’ll do both, but keep it simple.

```python
# Define the cost function J(w) = (w - 3)^2

def J(w):
    return (w - 3)**2

# Optional: exact derivative J'(w) = 2(w - 3)
def dJ_exact(w):
    return 2 * (w - 3)
```

It means:
- Take your current value of w.

- Subtract 3 → (w - 3).

- Square it → (w - 3)**2.

- J(3)=(3−3)^2 =0 → minimum cost

```python
# Visualize J(w)
w_vals = np.linspace(-2, 8, 200)
J_vals = J(w_vals)

plt.plot(w_vals, J_vals)
plt.axvline(3, color="red", linestyle="--", label="True minimum at w = 3")
plt.title("Cost Function J(w) = (w - 3)^2")
plt.xlabel("w")
plt.ylabel("J(w)")
plt.legend()
plt.grid(True)
plt.show()
```

<div style="background-color: lightblue; color: black; padding: 4px;">
    <h4> Gradient Descent on J(w) = (w - 3)²

#### Algorithm Steps (1D case)

1. Choose a starting value for w (e.g., w = 0)
2. Choose a **learning rate** (small positive number, e.g., 0.1)
3. Repeat:
   - Compute derivative \( J'(w) \)
   - Update:
     \[
     w_new = w_old - learning_rate * J'(w_old)
     \]

We move **against the gradient**, towards lower values of J(w).

```python
# Gradient descent on J(w) = (w - 3)^2

learning_rate = 0.1  # how big a step we take each time
n_iterations = 20    # how many updates we do

# Start from a random point, e.g., w = 0
w = 0.0


for i in range(n_iterations):
    cost = J(w)
    grad = dJ_exact(w)  # using exact derivative here 
    print("Step:", i, "| w:", w, "| cost:", cost, "| grad:", grad)    
    
    # Gradient descent update rule
    # move w in the opposite direction of the slope
    n_iterations
    w = w - learning_rate * grad

print("\nFinal w after gradient descent:", w)
```

#### To track all the changes, let's do this:

```python
learning_rate = 0.1
n_iterations = 20

# Start from a random point, e.g., w = 0
w = 0.0

# For tracking
w_history = []
J_history = []

for i in range(n_iterations):
    cost = J(w)
    grad = dJ_exact(w)  # using exact derivative here
    
    w_history.append(w)
    J_history.append(cost)
    
    print(f"Iter {i:2d} | w = {w:.4f} | J(w) = {cost:.4f} | J'(w) = {grad:.4f}")
    
    # Gradient descent update rule
    w = w - learning_rate * grad

print("\nFinal w after gradient descent:", w)
```

```python
# 8.2 Visualizing the path of gradient descent on the cost curve

plt.plot(w_vals, J_vals, label="J(w)")

plt.scatter(w_history, J_history, color="red", label="Gradient descent steps")

plt.plot(w_history, J_history, color="red", linestyle="--")

for i, (w_i, J_i) in enumerate(zip(w_history, J_history)):
    plt.text(w_i, J_i + 0.5, str(i), fontsize=8)

plt.axvline(3, color="green", linestyle="--", label="True minimum w = 3")
plt.title("Gradient Descent Path on J(w) = (w - 3)^2")
plt.xlabel("w")
plt.ylabel("J(w)")
plt.legend()
plt.grid(True)
plt.show()
```

#### Observation

- w starts at 0 and moves closer to 3 step by step
- J(w) (the cost) keeps **decreasing**
- If the learning rate is **too small** → slow movement  
- If the learning rate is **too large** → can overshoot or diverge

This is the **basic idea** of how ML algorithms adjust their parameters.

    <h3> Summary

You should now understand:

- A **derivative** is the slope (rate of change) of a function at a point.
- If the slope is **positive**, the function is going up as x increases.
- If the slope is **negative**, the function is going down as x increases.
- In ML, we use derivatives/gradients to know:
  - **How to change parameters** to reduce the loss.
- **Gradient descent**:
  - Starts with random parameters
  - Repeatedly moves in the **downhill direction**
  - Tries to reach a **minimum** of the loss function

These ideas will appear again later in:
- **Logistic Regression**
- **Neural Networks**
- **Deep Learning**

You don’t have to do heavy manual math.  
You just need to understand:  
 *“Gradients show the direction of change, and gradient descent walks downhill on the loss.”*
