```python
arr1D = np.array([1,  2,  3])
print(arr1D,"\n",arr1D.ndim)
```

```python
arr2D = np.array([[1,  2,  3],
                  [1,  2,  3]])

print(arr2D,"\n",arr2D.ndim)
```

```python
arr3D = np.array([[[1,  2,  3],[1,  2,  3]],
                  [[1,  2,  3],[1,  2,  3]]])

print(arr3D,"\n",arr3D.ndim)
```

```python
# ndim works on objects, and checks their dimensions

print(f"dimensions are: {arr1D.ndim}")

print(f"dimensions are: {arr2D.ndim}")

print(f"dimensions are: {arr3D.ndim}")


# using np.shape method, for checking the dimensions of an array!!!
# (rows, columns)

print(f"dimensions are: {np.shape(arr1D)}")

print(f"dimensions are: {np.shape(arr2D)}")

print(f"dimensions are: {np.shape(arr3D)}")

np.ndim(arr1D)
arr1D.shape

# size, how large array is, how many elements there are in the array


# 

arr1D.dtype

```
